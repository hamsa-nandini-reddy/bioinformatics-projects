```python
dna = "AAAACCCGGT"
complement_map = {
    'A': 'T',
    'T': 'A',
    'C': 'G',
    'G': 'C'
}

complement = ''
for base in dna:
    complement = complement + complement_map[base]
reverse_complement = complement[::-1]
print(reverse_complement)
```
