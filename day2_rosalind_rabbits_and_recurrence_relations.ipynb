```python
# FIBONACCI RABBITS
# n = number of months
# k = number of pairs produced by each adult pair per month

n = 5    # number of months
k = 3    # rabbit pairs produced per month

# We track two groups each month:
# young = newborn pairs (not yet adults)
# adult = pairs that can reproduce

young = 1    # start with 1 pair of baby rabbits
adult = 0    # no adults yet

# Loop through each month (we already set up month 1 above, so start from month 2)
for month in range(2, n + 1):
    # New babies = each adult pair produces k new pairs
    new_young = adult * k

    # This month's adults = last month's young (they grew up) + last month's adults (still alive)
    new_adult = young + adult

    # Update for next month
    young = new_young
    adult = new_adult

    print(f"Month {month}: {young} young pairs, {adult} adult pairs, {young + adult} total")

# Final answer = total pairs at month n
total = young + adult
print(f"\nAnswer: {total}")
```
